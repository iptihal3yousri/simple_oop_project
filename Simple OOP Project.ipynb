```python
class members:
    def __init__(self):
        self.__name = input("please, enter your name: ")
        self.__age = input("please, enter your age: ")
        self.__email = input("please, enter your email: ")
        self.__phone = input("plese enter your phone: ")
        self.__password = int(input("please enter your password: "))
    
    def _checkpassword(self):
        userpassword = int(input("please, enter your password: "))
        if userpassword == self.__password:
            return True
        else:
            return False

    def changepassword(self):
        if self._checkpassword() == True:
            newpassword = int(input("please enter the new password: "))
            self.__password = newpassword
        else:
            print("incorrect password")
            
    def _getName(self):
        return self.__name
    
    
    def changephone(self):
        if self._checkpassword() == True:
            newphone = input("please enter the new phone: ")
            self.__phone = newphone
        else:
            print("incorrect password")
    
    def changeEmail(self):
        if self._checkpassword() == True:
            newemail = input("please enter the new email: ")
            self.__email = newemail
        else:
            print("incorrect password")
            
    def getname(self):
        return self.__name
            
    def _display(self, mem):
        print("The ", mem, " name : ", self.__name,"\nThe ", mem," age : ", self.__age,"\nThe ", mem, " email : ", self.__email,"\nThe ", mem," Phone : ", self.__phone)
```

```python
class student(members):
     def __init__(self):
        members.__init__(self)
        self.__currentCourses = []
        self.__degree = []
        self.__StudentID = input("enter your ID : ")
        self.level = input("please enter your level: ")
        print("please enter the six courses names and grades")
        for i in range(6):
            courseName = input("please enter the course name: ")
            coursedegree = int(input("please enter the course grade: "))
            self.__currentCourses.append(courseName)
            self.__degree.append(coursedegree)
        self.__percentage = (sum(self.__degree)/600)*100
        
    
     def __checkStudenID(self):
        SID = input("please, enter your ID: ")
        if SID == self.__StudentID:
            return True
        else:
            return False
    
     def getpercentage(self):
        if members._checkpassword(self) == True:
            print("Your percentage is: ", self.__percentage,"%")
        else:
            print("incorrect password")
   

     def display(self):
        if self.__checkStudenID() == True:
            members._display(self, "Student")
            print("The Student level : ", self.level, "\nThe Student percentage : ", self.__percentage)
            print("the current courses with their degrees : ")
            for i in range(6):
                print(self.__currentCourses[i], " : ", self.__degree[i])
        else:
            print("incorrect ID")
        
```
